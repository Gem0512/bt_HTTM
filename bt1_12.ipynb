```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

data = {
    'diện_tích': [50, 60, 65, 70, 75, 80, 85],
    'giá_nhà': [2.5, 3, 3.5, 3.8, 4, 4.5, 5]
}

df = pd.DataFrame(data)
```

```python
# Tạo mô hình hồi quy tuyến tính
model = LinearRegression()

# Đào tạo mô hình trên dữ liệu
X = df[['diện_tích']]
y = df['giá_nhà']
model.fit(X, y)
```

```python
# Các diện tích cần dự đoán
new_areas = np.array([55, 68, 76, 90]).reshape(-1, 1)

# Dự đoán giá nhà cho các diện tích mới
predicted_prices = model.predict(new_areas)

# Tạo DataFrame kết quả
predictions = pd.DataFrame({'Diện_tích': new_areas.flatten(), 'Giá_nhà_dự_đoán': predicted_prices})

print(predictions)


plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, model.predict(X), color='red', label='Linear Regression')
plt.scatter(new_areas, predicted_prices, color='green', label='Dự đoán')
plt.xlabel('Diện tích (m2)')
plt.ylabel('Giá nhà (tỷ)')
plt.title('Dự đoán giá nhà bằng Linear Regression')
plt.legend()
plt.show()
```
