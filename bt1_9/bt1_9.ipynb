```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data from CSV
data = pd.read_csv('data.csv')

# Create the categorical plot
plt.figure(figsize=(8, 6))
plot = sns.catplot(
    data=data,
    x='group', y='license', hue='gender',
    kind='bar', ci=None
)

# Set titles and labels
plot.set(title='Proportions of Men and Women with Driver\'s License by Group')
plt.xlabel('Group')
plt.ylabel('Proportion')
plt.legend(title='Gender')

plt.show()
```
